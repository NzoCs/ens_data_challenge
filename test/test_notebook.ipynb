```python
import pandas as pd
import numpy as np
```

```python
# Exemple de DataFrame
df = pd.DataFrame({
    'A': ['x', 'x', 'y', 'z', 'z', 'z'],
    'B': ['foo', 'bar', 'foo', 'foo', 'bar', 'bar']
})
```

```python
#
categorical_cols = ['A', 'B']
thresh = 0.3  # Seuil de frÃ©quence minimale

mask = np.ones(len(df), dtype=bool)
for col in categorical_cols:
    freq = df[col].map(df[col].value_counts(normalize=True))
    mask &= freq > thresh

df[np.logical_not(mask)] = "other"
print(df)
```
