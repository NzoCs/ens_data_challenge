```python
from ens_data_challenge.preprocess import Preprocessor
from ens_data_challenge.globals import (
    TEST_CLINICAL_DATA_PATH, 
    TEST_MOLECULAR_DATA_PATH, 
    TRAIN_CLINICAL_DATA_PATH, 
    TRAIN_MOLECULAR_DATA_PATH
)
import pandas as pd
```

```python
train_clinical = pd.read_csv(TRAIN_CLINICAL_DATA_PATH)
train_molecular = pd.read_csv(TRAIN_MOLECULAR_DATA_PATH)
test_clinical = pd.read_csv(TEST_CLINICAL_DATA_PATH)
test_molecular = pd.read_csv(TEST_MOLECULAR_DATA_PATH)
preprocessor = Preprocessor()
```

```python
cyto_struct_train, cyto_struct_test = preprocessor.fit(train_clinical, train_molecular, test_clinical, test_molecular)
```

```python
test_clinical_processed, test_molecular_processed, test_cyto_processed = preprocessor.transform(test_clinical, test_molecular, cyto_struct_test)
train_clinical_processed, train_molecular_processed, train_cyto_processed = preprocessor.transform(train_clinical, train_molecular, cyto_struct_train)
```
