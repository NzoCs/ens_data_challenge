```python
from ens_data_challenge.preprocess.preprocessor import Preprocessor
from ens_data_challenge.feature_engineering import FeatureEngineerHelper
from ens_data_challenge.types import CytoColumns, MolecularColumns, CytoStructColumns
from ens_data_challenge.globals import TRAIN_CLINICAL_DATA_PATH, TRAIN_MOLECULAR_DATA_PATH, TEST_CLINICAL_DATA_PATH, TEST_MOLECULAR_DATA_PATH, TRAIN_TARGET_PATH

import pandas as pd

clinical_data_test = pd.read_csv(TEST_CLINICAL_DATA_PATH)
molecular_data_test = pd.read_csv(TEST_MOLECULAR_DATA_PATH)

clinical_data_train = pd.read_csv(TRAIN_CLINICAL_DATA_PATH)
molecular_data_train = pd.read_csv(TRAIN_MOLECULAR_DATA_PATH)

target_data = pd.read_csv(TRAIN_TARGET_PATH)

# For test purposes, use the same data for train and val
train_targets = target_data
val_targets = target_data


pe = Preprocessor()

cyto_struct_train, cyto_struct_val = pe.fit(
    clinical_data_train=clinical_data_train, 
    molecular_data_train=molecular_data_train,
    clinical_data_test=clinical_data_test,
    molecular_data_test=molecular_data_test,
)

clinical_train_clean, molecular_train_clean, cytogenetic_train_clean, train_targets_clean = pe.transform(
    clinical_data_train, molecular_data_train, cyto_struct_train, train_targets
)

clinical_val_clean, molecular_val_clean, cytogenetic_val_clean, val_targets_clean = pe.transform(
    clinical_data_test, molecular_data_test, cyto_struct_val, val_targets
)
```

```python
fe = FeatureEngineerHelper(
    clinical_data_train=clinical_train_clean, 
    molecular_data_train=molecular_train_clean,
    cytogenetic_data_train=cytogenetic_train_clean,
    clinical_data_test=clinical_val_clean,
    molecular_data_test=molecular_val_clean,
    cytogenetic_data_test=cytogenetic_val_clean,
    target_data_train=train_targets,
)

fe.as_type(
    type_dict={
        CytoColumns.IS_NORMAL.value: 'int32',
    },
    data_type='clinical'
)

fe.add_mol_encoding(
    col='GENE',
    method='vaf_score',
    apply_effect_weighting=False
)

fe.add_mol_encoding(
    col='PATHWAY',
    method='constant',
    apply_effect_weighting=False
)

fe.add_mol_encoding(
    col='EFFECT',
    method='constant',
    apply_effect_weighting=False
)

fe.encode_categorical(
    categorical_cols=[
        CytoColumns.MDS_IPSS_R_CYTO_RISK.value,
        CytoColumns.AML_ELN_2022_CYTO_RISK.value,
        CytoColumns.CLL_CYTO_RISK.value,
        CytoColumns.MM_RISS_CYTO_RISK.value,
    ],
    data_type='clinical'
)

fe.encode_categorical(
    categorical_cols=[
        MolecularColumns.REF.value,
        MolecularColumns.ALT.value,
        MolecularColumns.CHR.value
    ],
    data_type='molecular'
)

fe.encode_categorical(
    categorical_cols=[
        CytoStructColumns.MUTATION_TYPE.value,
        CytoStructColumns.ARM.value,
        CytoStructColumns.SEX_CHROMOSOMES.value,
        CytoStructColumns.END_ARM.value,
        CytoStructColumns.START_ARM.value,
    ],
    data_type='cytogenetic'
)
```

```python
clinical_train_processed, clinical_val_processed = fe.get_clinical_data()
molecular_train_processed, molecular_val_processed = fe.get_molecular_data()
cytogenetic_train_processed, cytogenetic_val_processed = fe.get_cytogenetic_data()
```
